```python
wikidata_root_url = 'https://www.wikidata.org/w/api.php?format=json'
wikidata_get_entity_url = wikidata_root_url + '&action=wbgetentities&sites=dewiki|frwiki|itwiki|enwiki&languages=de|fr'

def get_wikidata_id(name):
    req = requests.get(wikidata_get_entity_url + '&titles=' + name)
    
    if req.status_code != 200:
      raise Exception('Request failed')
    
    ids = req.json()['entities'].keys()
    
    return [id for id in ids if not id.startswith('-')]

get_wikidata_id('École polytechnique fédérale de Lausanne')
```

```python
# We are only interested in the P625 (geolocation) and P159 (headquater location) properties
wikidata_get_location_claim_url = wikidata_root_url + '&action=wbgetclaims&property=P625'
wikidata_get_hq_claim_url = wikidata_root_url + '&action=wbgetclaims&property=P159'

def get_claims(url):
    req = requests.get(url)
    
    if req.status_code != 200:
      raise Exception('Request failed')
    
    return req.json()['claims']

def get_wikidata_headquater_location(id):
    claims = get_claims(wikidata_get_hq_claim_url + '&entity=' + id)
    
    if 'P159' in claims:
        claim = claims['P159'][0]
        
        if 'qualifiers' in claim and 'P625' in claim['qualifiers']:
            return claim['qualifiers']['P625'][0]['datavalue']['value']

def get_wikidata_geolocation(id):
    claims = get_claims(wikidata_get_location_claim_url + '&entity=' + id)
    
    if 'P625' in claims:
        return claims['P625'][0]['mainsnak']['datavalue']['value']
    
    return get_wikidata_headquater_location(id)

get_wikidata_geolocation('Q262760')
```

```python
# Remove abbreviations and initials in the data

replacements = {
    'Staatsunabh.': 'Staatsunabhängige',
    'Pädag.': 'Pädagogische',
    'Schweiz.': 'Schweizer',
    'rech.': 'recherche',
    'Eidg.': 'Eidgenössische',
    'Inst.': 'Institut',
    'HOS': 'Haute école spécialisée',
    'EPF ': 'École polytechnique fédérale de ',
    'Microtech.': 'Microtechnology',
    'Physikal.': 'Physikalisch',
    'Meteorolog.': 'Meteorologisches',
    'Sozialwissensch.': 'Sozialwissenschaft',
    
    'Forschungskommission SAGW': 'Schweizerische Akademie der Geistes- und Sozialwissenschaften ',
    'Forschungsinstitut für Opthalmologie' :'Institut de Recherche en Ophtalmologie',
    'BE, JU, NE': 'BEJUNE',    
    'Wald,Schnee,Land': 'Wald, Schnee und Landschaft',
    'Eidg. Material und Prüfungsanstalt': 'EMPA',   
}

def get_cleaned_name(name):
    name = name.split(' - ', 2)[0].split(' (', 2)[0]
    for key in replacements:
        name = name.replace(key, replacements[key])
    return name

get_cleaned_name('Eidg. Material und Prüfungsanstalt')
```

```python
fr_identifiers = [
    'école', 'université',
    ' de ', ' du ',
    'universitaire',
    'idiap research institute', # located in Martigny
    'swiss center for electronics and microtechnology', # located in Neuchâtel
    'swiss institute of bioinformatics' # located in Lausanne
]
it_identifiers = [
    'instituto', 'università', ' di ', 'italiana',
    'franklin university', # located in Ticino
    'ente ospedaliero cantonale' # located in Ticino
]
de_identifiers = [
    'schule', 'kommission', ' für',
    'universität', 'weitere ', 'stiftung',
    'schweizer', 'spital', 'forschung',
    'anstalt', 'zürich', 'davos', 'biotechnologie',
    'ao research institute', #located in Davos
    'paul scherrer institut', # located in Aargau
    'friedrich miescher institute', # located in Basel
    'empa', # located in Dübendorf
]

def get_language(name):
    name = name.lower()

    for ident in it_identifiers:
        if ident in name:
            return 'it'
    
    for ident in fr_identifiers:
        if ident in name:
            return 'fr'
    
    for ident in de_identifiers:
        if ident in name:
            return 'de'
    
    print(name)


grants['Language'] = grants.apply(get_language, axis=1)
#university_languages
grants.groupby('Language')['Approved Amount'].sum().sort_values()
```

```python
canton_cache = {'Swiss Institute of Bioinformatics': 'VD',
                    'Schweizer Kompetenzzentrum Sozialwissenschaft': 'VD',
                    'Forschungsinstitut für biologischen Landbau' :'AG',
                    'Istituto Svizzero di Roma': str(None),
                    'Institut de Hautes Etudes Internat. et du Dév': 'GE'}

def get_canton(row):
    cleaned_name = get_cleaned_name(row.University) 
    
    if cleaned_name in canton_cache:
        return canton_cache[cleaned_name]
    
    req = requests.get('https://maps.googleapis.com/maps/api/place/textsearch/json?query=' + cleaned_name + '&key=' + key)
    
    if req.status_code != 200:
      raise Exception('Request failed')
    
    json_res = req.json()
    addr = json_res['results'][0]['formatted_address'] if len(json_res['results']) > 0 else None
    
    canton = None
    if addr is not None:
        req = requests.get('https://maps.googleapis.com/maps/api/geocode/json?address=' + addr + '&sensor=false')
    
        if req.status_code != 200:
            raise Exception('Request failed')

        json_res = req.json()
        address_comps = json_res['results'][0]['address_components'] if len(json_res['results']) > 0 else None
        
        if address_comps is not None:
            canton =[addr_comp['short_name'] for addr_comp in address_comps if 'administrative_area_level_1' in addr_comp['types']]
            canton = canton[0] if len(canton) > 0 else None
    
    print(cleaned_name + ': address: ' + str(addr) + ' canton: ' + str(canton))
    
    canton_cache[cleaned_name] = canton
    return canton
```

```python
grants['Canton'] = grants.apply(get_canton, axis=1)
grants
```

```python
grant_sums = grants.groupby('Canton')['Approved Amount'].sum().sort_values(ascending=False)
grant_sums = pd.DataFrame({'Canton': grant_sums.index, 'Funding': grant_sums.values})
grant_sums.loc[len(grant_sums)] = ('UR', 0.0)
grant_sums.loc[len(grant_sums)] = ('OW', 0.0)
grant_sums.loc[len(grant_sums)] = ('NW', 0.0)
grant_sums.loc[len(grant_sums)] = ('GL', 0.0)
grant_sums.loc[len(grant_sums)] = ('BL', 0.0)
grant_sums.loc[len(grant_sums)] = ('AR', 0.0)
grant_sums.loc[len(grant_sums)] = ('AI', 0.0)
grant_sums
```
