```python
import folium
import pandas as pd
import requests
```

```python
grants = pd.read_csv('data/P3_GrantExport.csv', sep=';')
grants
```

```python
# Drop entries not assignable to a university
grants = grants[(grants.University != 'Nicht zuteilbar - NA')]
# Drop entries without an assigned university
grants = grants[(pd.notnull(grants.University))]

grants['Approved Amount'] = pd.to_numeric(grants['Approved Amount'], errors='coerce')
grants
```

```python
universities = grants.groupby('University')['Approved Amount'].sum().sort_values().keys()
```

```python
universities = grants['University'].unique()
universities
```

```python
wikidata_root_url = 'https://www.wikidata.org/w/api.php?format=json'
wikidata_get_entity_url = wikidata_root_url + '&action=wbgetentities&sites=dewiki|frwiki|itwiki|enwiki&languages=de|fr'

def get_wikidata_id(name):
    req = requests.get(wikidata_get_entity_url + '&titles=' + name)
    
    if req.status_code != 200:
      raise Exception('Request failed')
    
    ids = req.json()['entities'].keys()
    
    return [id for id in ids if not id.startswith('-')]

get_wikidata_id('École polytechnique fédérale de Lausanne')
```

```python
# We are only interested in the P625 (geolocation) and P159 (headquater location) properties
wikidata_get_location_claim_url = wikidata_root_url + '&action=wbgetclaims&property=P625'
wikidata_get_hq_claim_url = wikidata_root_url + '&action=wbgetclaims&property=P159'

def get_claims(url):
    req = requests.get(url)
    
    if req.status_code != 200:
      raise Exception('Request failed')
    
    return req.json()['claims']

def get_wikidata_headquater_location(id):
    claims = get_claims(wikidata_get_hq_claim_url + '&entity=' + id)
    
    if 'P159' in claims:
        claim = claims['P159'][0]
        
        if 'qualifiers' in claim and 'P625' in claim['qualifiers']:
            return claim['qualifiers']['P625'][0]['datavalue']['value']

def get_wikidata_geolocation(id):
    claims = get_claims(wikidata_get_location_claim_url + '&entity=' + id)
    
    if 'P625' in claims:
        return claims['P625'][0]['mainsnak']['datavalue']['value']
    
    return get_wikidata_headquater_location(id)

get_wikidata_geolocation('Q262760')
```

```python
# Remove abbreviations and initials in the data

replacements = {
    'Staatsunabh.': 'Staatsunabhängige',
    'Pädag.': 'Pädagogische',
    'Schweiz.': 'Schweizer',
    'rech.': 'recherche',
    'Eidg.': 'Eidgenössische',
    'Inst.': 'Institut',
    'HOS': 'Haute école spécialisée',
    'EPF ': 'École polytechnique fédérale de ',
    
    'Wald,Schnee,Land': 'Wald, Schnee und Landschaft'
}

def get_cleaned_name(name):
    name = name.split(' - ', 2)[0].split(' (', 2)[0]
    for key in replacements:
        name = name.replace(key, replacements[key])
    return name

get_cleaned_name('EPF Lausanne')
```

```python
wikidata_ids = {name: get_wikidata_id(get_cleaned_name(name)) for name in universities}
wikidata_ids
```

```python
wikidata_locations = {name: (get_wikidata_geolocation(wikidata_ids[name][0]) if len(wikidata_ids[name]) > 0 else None) for name in wikidata_ids}
wikidata_locations
```

```python
count = 0

for entry in wikidata_locations.values():
    if entry != None:
        count += 1
        

count
```
